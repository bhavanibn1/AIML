one sample test

```python
import numpy as np
from scipy import stats
d=pd.read_csv('/content/internet dataset.zip')
urbanrate=np.array([24.04,46.72,65.22,88.92,56.07])
t_stat,p_value= stats.ttest_1samp(urbanrate,popmean=7.2)
print( f"one sample t-test:{t_stat},p_value:{p_value}")
```

```python
chisquare
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
data = pd.read_csv('/content/internet dataset.zip')
contingency_table = pd.crosstab(data['country'], data['urbanrate'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f"Chi-square Test of Independence: chi2={chi2},p={p},dof={dof},expected={ex}")
```

```python
import pandas as pd
from scipy.stats import f_oneway
d=pd.read_csv('/content/internet dataset.zip')
groups = [d['country'], d['urbanrate'], d['incomeperperson']]
f_stat, p_value = f_oneway(*groups)
print(f"F-value:{f_stat},p-value:{p_value}")
```

wilcoxon

```python
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon
d=pd.read_csv('/content/internet dataset.zip')
data={
    'before':[24.04,46.72,65.22,88.92,56.07],
    'after':[30,40,35,20,50]
}
df=pd.DataFrame(data)
stat,p_value=wilcoxon(df['before'],df['after'])
print(f"Wilcoxon signed-rank statistic:{stat},p-value:{p_value}")
```

```python
import pandas as pd
from scipy.stats import kruskal
import numpy as np
d=pd.read_csv('/content/internet dataset.zip')
d={
'group1':np.random.normal(loc=0,scale=1,size=10),
'group2':np.random.normal(loc=0,scale=1,size=10),
'group3':np.random.normal(loc=0,scale=1,size=10)
}
f_stat,p_value=kruskal(d['group1'],d['group2'],d['group3'])
print(f"kruskal-wallis H statistic:{f_stat},p-value:{p_value}")
```
